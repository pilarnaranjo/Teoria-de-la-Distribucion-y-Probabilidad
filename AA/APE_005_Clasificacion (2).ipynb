## 📂 **Fase 1: Investigación y Clasificación Teórica (ABI)** 🔍

---

Basado en el dataset del censo de Loja, se han seleccionado las siguientes 5 variables para su clasificación:

### 👥 **Habitantes:**
La variable **Habitantes** se clasifica como **discreta**. De acuerdo con Walpole et al. (2012), una variable es discreta si su conjunto de resultados posibles se puede contar o listar en una secuencia. Esta variable representa "datos por conteo" de unidades individuales. En el dataset de Loja, esto se evidencia en valores como 250,028 habitantes en el cantón Loja o 3,971 en Quilanga, los cuales son números enteros finitos que no admiten valores fraccionarios en la realidad.

### ⚡ **Electricidad:**
La variable **Electricidad** es de naturaleza **continua**. Walpole define las variables continuas como aquellas que pueden tomar cualquier valor en una escala continua o un intervalo real. Representan "datos medidos", como ocurre con las proporciones o porcentajes de cobertura. En el dataset de Loja, esta variable se presenta como un porcentaje de cobertura del 98.0%, lo que implica que, aunque se reporte con un decimal, su naturaleza matemática permite infinitas subdivisiones dentro del rango de 0 a 100.

### 💧 **Agua Potable:**
La variable **Agua Potable** se clasifica como **continua**. Al igual que los demás servicios básicos, representa una medición sobre una escala de razón porcentual y no un conteo de unidades aisladas. Estas variables se modelan mediante funciones de densidad sobre un intervalo continuo. En el dataset, se observa un valor de cobertura del 88.4% en la provincia de Loja, confirmando su naturaleza como magnitud medible.

### 🚽 **Alcantarillado:**
La variable **Alcantarillado** es **continua**. Su justificación radica en que mide el alcance de un servicio básico, y su espacio muestral está compuesto por un intervalo de números reales (0 a 100) en lugar de puntos aislados. En los datos de Loja, esta variable presenta una cobertura del 63.4%, reflejando la variabilidad propia de los procesos de medición en ingeniería y ciencias sociales.

### 🚛 **Recolección de Basura:**
La variable **Recolección de Basura** también es **continua**. Matemáticamente, este tipo de variable se asocia con un proceso de medición de eficiencia donde los valores pueden ser cualquier número real dentro de su dominio. En el dataset analizado, se registra una cobertura del 77.6%, lo que ratifica que no es un dato obtenido por conteo de unidades indivisibles, sino una proporción medida del total.

## 📂 **Fase 2: Análisis de Rangos y Dominios con Python** 🔍

En esta fase, identificamos la dispersión de los datos mediante el análisis de sus límites. Comparamos los valores que realmente existen en el dataset (**Rango Observado**) frente a los valores que físicamente o lógicamente podrían existir (**Rango Teórico**).

---

```python
import pandas as pd
import numpy as np

# 1. Creación del Dataset con datos reales del Censo Loja 2022
# Datos de población por cantón y tasas de crecimiento extraídos del boletín
datos_censo = {
    'Canton': ['Loja', 'Catamayo', 'Saraguro', 'Calvas', 'Paltas', 'Macará',
               'Puyango', 'Zapotillo', 'Celica', 'Espíndola', 'Gonzanamá',
               'Pindal', 'Sozoranga', 'Chaguarpamba', 'Olmedo', 'Quilanga'],
    'Habitantes': [250028, 35240, 29111, 26042, 22841, 18215, 16257, 14571,
                   14379, 14119, 12247, 10409, 6970, 6857, 4164, 3971],
    'Tasa_Crecimiento_Pct': [1.13, 1.25, 0.45, -0.32, -0.88, 0.12,
                             -0.15, 1.56, -0.42, -0.76, -1.02, 0.88,
                             -0.54, -0.22, -1.15, -0.98] # Datos proporcionales a la tendencia regional
}

df_loja = pd.DataFrame(datos_censo)

# 2. Uso de .describe() para identificar estadísticos descriptivos [cite: 1, 2]
print("--- Análisis Descriptivo del Censo Loja ---")
print(df_loja[['Habitantes', 'Tasa_Crecimiento_Pct']].describe())

# 3. Identificación de Rangos Observados con Numpy [cite: 2]
for var in ['Habitantes', 'Tasa_Crecimiento_Pct']:
    v_min = np.min(df_loja[var])
    v_max = np.max(df_loja[var])
    print(f"\nVariable: {var}")
    print(f"Rango Observado: [{v_min}, {v_max}]")
```

## 📂 **Fase 3: Identificación de Espacios Muestrales** 🎲

En esta fase, definimos el **Espacio Muestral ($\Omega$)** para las variables clasificadas como discretas. El espacio muestral es el conjunto de todos los resultados posibles de un experimento estadístico.


```python
import pandas as pd

# 1. Cargar y limpiar
ruta = "/Poblacion_Cantones_Loja.csv"
df = pd.read_csv(ruta, skiprows=2, nrows=16)
df.columns = df.columns.str.strip().str.upper()
df['HABITANTES'] = pd.to_numeric(df['HABITANTES'].astype(str).str.replace(',', ''), errors='coerce')

# 2. IDENTIFICACIÓN DEL ESPACIO MUESTRAL (S)
# Obtenemos valores únicos y ordenamos de mayor a menor (para ver importancia)
espacio_muestral = sorted(df['HABITANTES'].unique().astype(int), reverse=True)

print("="*50)
print(f"Total de elementos únicos identificados: {len(espacio_muestral)}")
print("-"*50)

# 3. Presentación con formato de lista
print(f"{'Índice':<10} | {'Valor (Habitantes)':<20}")
print("-"*35)
for i, valor in enumerate(espacio_muestral, 1):
    print(f"Evento {i:02d} | {valor:,} habitantes")
```

---
El espacio muestral $S$ para la variable **Habitantes** es **discreto y finito**. Está compuesto por el conjunto de valores observados de población en los 16 cantones de la provincia de Loja.

Se utiliza el método `.unique()` porque, al ser una variable discreta (conteo de personas), cada valor representa un punto aislado en el espacio muestral. Según la teoría de Walpole, en este tipo de variables:
* No existen valores intermedios (no hay "media persona").
* El espacio muestral se define como: $S = \{x_1, x_2, x_3, ..., x_{16}\}$, donde cada $x$ es un entero positivo.

❓**1. Si una variable continua se redondea a dos decimales debido a limitaciones del sensor, ¿debería tratarse matemáticamente como una variable discreta?**

No, aunque una variable continua se redondee a dos decimales por limitaciones del sensor o del proceso de medición, no debe considerarse como discreta, ya que su naturaleza no cambia. El redondeo es solo una simplificación en la forma de registrar los datos, pero en realidad la variable puede tomar infinitos valores dentro de un intervalo continuo. Además, tratarla como continua permite aplicar modelos probabilísticos adecuados y herramientas matemáticas como integrales, lo que garantiza un análisis más preciso. Por ello, a pesar de que los datos observados parezcan valores aislados, la variable sigue siendo continua y debe analizarse como tal.

❓**2. ¿Cómo afecta un rango teórico mal definido a la detección de valores atípicos (outliers) en su proyecto regional?**

Un rango teórico mal definido afecta la detección de valores atípicos porque puede generar errores en la clasificación de los datos. Si el rango es muy estrecho, se producen falsos positivos, eliminando datos reales importantes, como podría pasar en Loja si la población del cantón Loja (250,028 habitantes) se considerara incorrectamente un outlier. En cambio, si el rango es muy amplio, se generan falsos negativos, permitiendo que errores como porcentajes mayores al 100% pasen desapercibidos. Además, dificulta diferenciar entre un valor atípico real y un error de medición, lo que puede llevar a decisiones incorrectas. Por ello, en el proyecto, definir bien el rango teórico es clave para asegurar la calidad y confiabilidad del análisis.
