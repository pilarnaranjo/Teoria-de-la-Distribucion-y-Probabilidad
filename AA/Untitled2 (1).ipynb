###📊 **Desarrollo Algorítmico**

```python
import pandas as pd
import numpy as np

# 1. Cargar dataset
df = pd.read_csv("datos_loja.csv")

# 2. Variable
datos = df["Sin_Alcantarillado"].values


# 3. Esperanza (media) manual
def calcular_esperanza(datos):
    suma = 0
    n = 0

    for x in datos:
        suma = suma + x
        n = n + 1

    return suma / n


# 4. Varianza manual
def calcular_varianza(datos, insesgada=True):
    media = calcular_esperanza(datos)

    suma = 0
    n = 0

    for x in datos:
        suma = suma + (x - media) ** 2
        n = n + 1

    if insesgada:
        return suma / (n - 1)
    else:
        return suma / n


# 5. Cálculo manual
media_manual = calcular_esperanza(datos)
var_manual = calcular_varianza(datos, insesgada=True)

# 6. Validación con NumPy
media_numpy = np.mean(datos)
var_numpy = np.var(datos, ddof=1)

# 7. Resultados
print("=== RESULTADOS ===\n")

print(f"Media manual: {media_manual}")
print(f"Media NumPy:  {media_numpy}\n")

print(f"Varianza manual: {var_manual}")
print(f"Varianza NumPy:  {var_numpy}\n")

print("Diferencias:")
print(f"Media: {abs(media_manual - media_numpy)}")
print(f"Varianza: {abs(var_manual - var_numpy)}")
```
