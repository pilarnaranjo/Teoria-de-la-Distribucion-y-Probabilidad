🧪 **TAREA 1: ESPACIOS MUESTRALES**

```python
import itertools

# Moneda
moneda = ['C', 'S']

omega_1 = list(itertools.product(moneda, repeat=1))
omega_2 = list(itertools.product(moneda, repeat=2))
omega_3 = list(itertools.product(moneda, repeat=3))

print("1 lanzamiento:", omega_1)
print("2 lanzamientos:", omega_2)
print("3 lanzamientos:", omega_3)

# Dados
dado = [1, 2, 3, 4, 5, 6]

omega_dado1 = list(itertools.product(dado, repeat=1))
omega_dado2 = list(itertools.product(dado, repeat=2))

print("\n1 dado:", len(omega_dado1))
print("2 dados:", len(omega_dado2))
```

🎯 **TAREA 2: EVENTOS Y PROBABILIDAD CLÁSICA**

```python
import itertools

dado = [1, 2, 3, 4, 5, 6]
omega_dado2 = list(itertools.product(dado, repeat=2))

# Eventos
evento_A = [r for r in omega_dado2 if sum(r) == 7]
evento_B = [r for r in omega_dado2 if all(x % 2 == 0 for x in r)]
evento_C = [r for r in omega_dado2 if 6 in r]

def prob(evento, espacio):
    return len(evento) / len(espacio)

P_A = prob(evento_A, omega_dado2)
P_B = prob(evento_B, omega_dado2)
P_C = prob(evento_C, omega_dado2)

print("P(A):", P_A)
print("P(B):", P_B)
print("P(C):", P_C)

# Unión e intersección
inter_AB = [r for r in evento_A if r in evento_B]
union_AB = list(set(evento_A + evento_B))

P_A_inter_B = len(inter_AB) / len(omega_dado2)
P_A_union_B = len(union_AB) / len(omega_dado2)

print("\nA ∩ B:", inter_AB)
print("P(A ∩ B):", P_A_inter_B)
print("P(A ∪ B):", P_A_union_B)
print("Verificación:", P_A + P_B - P_A_inter_B)
```

🎲 **TAREA 3: SIMULACIÓN (EMPÍRICA)**

```python
import numpy as np

def simular_dados(n):
    resultados = []
    for _ in range(n):
        d1 = np.random.randint(1, 7)
        d2 = np.random.randint(1, 7)
        resultados.append((d1, d2))
    return resultados

# Simulación
for n in [100, 1000, 10000]:
    sim = simular_dados(n)
    conteo = sum(1 for r in sim if sum(r) == 7)
    print(f"n={n} -> P(suma=7) = {conteo/n:.4f}")
```

📌 **TAREA 4: PROBABILIDAD CONDICIONAL E INDEPENDENCIA**

```python
import itertools
import numpy as np

# Espacio muestral (2 dados)
dado = [1, 2, 3, 4, 5, 6]
omega_dado2 = list(itertools.product(dado, repeat=2))

# Evento A: suma = 7
evento_A = [r for r in omega_dado2 if sum(r) == 7]

# Evento B: ambos pares
evento_B = [r for r in omega_dado2 if all(x % 2 == 0 for x in r)]

# Probabilidad clásica
def prob(evento, espacio):
    return len(evento) / len(espacio)

P_A = prob(evento_A, omega_dado2)
P_B = prob(evento_B, omega_dado2)

# Intersección
interseccion_AB = [r for r in evento_A if r in evento_B]
P_A_inter_B = len(interseccion_AB) / len(omega_dado2)

# Probabilidad condicional P(A|B)
if P_B > 0:
    P_A_dado_B = P_A_inter_B / P_B
else:
    P_A_dado_B = 0

# Independencia
producto = P_A * P_B

print("Evento A:", evento_A)
print("Evento B:", evento_B)
print("A ∩ B:", interseccion_AB)

print("\nP(A):", P_A)
print("P(B):", P_B)
print("P(A ∩ B):", P_A_inter_B)
print("P(A|B):", P_A_dado_B)

print("\nP(A)*P(B):", producto)

if abs(P_A_inter_B - producto) < 0.0001:
    print("✔ Eventos INDEPENDIENTES")
else:
    print("✘ Eventos NO independientes")
```
