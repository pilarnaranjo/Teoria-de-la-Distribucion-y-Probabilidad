```python
def cal_bayes (prob_bdado_a, prob_a, prob_b):
  prob_evento = prob_bdado_a * prob_a / prob_b
  return prob_evento

# DATOS: P(Cáncer | Síntoma)
p_cancer = 1 / 100000
p_sintoma = 11 / 100000
p_sintoma_dado_cancer = 1 / 1

resultado = cal_bayes(p_sintoma_dado_cancer, p_cancer, p_sintoma)

print(f"Caso: VERDADERO POSITIVO")
print(f"Probabilidad: {resultado * 100:.2f}%")
```
